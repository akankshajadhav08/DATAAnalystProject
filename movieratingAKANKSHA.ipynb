```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configure plot style
sns.set(style='whitegrid')
```

```python
# Load Dataset
```

```python
import pandas as pd

# Correct the path (single set of quotes and raw string)
df = pd.read_csv(r"D:\Backup\Downloads\archive\imdb_top_1000.csv")

print("First 5 Rows of Dataset:")
print(df.head())
```

```python
# Data Overview
# ------------------------------------------
print("\nDataset Information:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Data Cleaning
```

```python
# Remove missing values
df.dropna(inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Average IMDb Rating
# ------------------------------------------
avg_rating = df['IMDB_Rating'].mean()
print(f"\nAverage IMDb Rating: {avg_rating:.2f}")
```

```python
# Distribution of IMDb Ratings
```

```python
# ------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['IMDB_Rating'], bins=10, kde=True, color='orange')
plt.title('Distribution of IMDb Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

```python
# Genre-wise Analysis
```

```python
# Top Genres by Count
plt.figure(figsize=(12,6))
df['Genre'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

```python
# Average Rating per Genre
genre_rating = df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
genre_rating.plot(kind='bar', color='green')
plt.title('Average IMDb Rating per Genre')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# Year-wise IMDb Rating Trend
# ------------------------------------------
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Released_Year', y='IMDB_Rating', marker='o')
plt.title('IMDb Rating Trend by Release Year')
plt.xlabel('Year')
plt.ylabel('IMDb Rating')
plt.tight_layout()
plt.show()
```

```python
# Votes vs IMDb Ratings
# ------------------------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(x='No_of_Votes', y='IMDB_Rating', data=df)
plt.title('Votes vs IMDb Rating')
plt.xlabel('Number of Votes')
plt.ylabel('IMDb Rating')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

# Select only numeric columns before computing correlation
numeric_df = df.select_dtypes(include='number')

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Conclusion
# ------------------------------------------
print("\nKey Insights:")
print(f"- Average IMDb rating: {avg_rating:.2f}")
print("- Genre-wise and year-wise rating trends plotted.")
print("- Votes positively correlated with ratings (as seen in scatter plot).")

print("\n Analysis Complete!")
```
